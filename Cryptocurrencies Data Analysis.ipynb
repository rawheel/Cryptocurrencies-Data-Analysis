```python
import requests 
import pandas as pd
import time
```

```python
def last_week_price(symbol,exchange="bitfinex", after ='2020-09-01'):
    #api ="https://api.cryptowat.ch/markets/kraken/btcusd/price"
    api ='https://api.cryptowat.ch/markets/{exchange}/{symbol}usd/ohlc'.format(symbol=symbol,exchange=exchange)

    resp=requests.get(api,params={'periods' : '3600', 'after':str(int(pd.Timestamp(after).timestamp()))})
    
    
    resp.raise_for_status()
    data = resp.json()
    
    df = pd.DataFrame(data['result']['3600'],columns=['CloseTime','OpenPrice','HigherPrice','LowPrice','ClosePrice','Volume','NA'])
    print(df)
    df['CloseTime']=pd.to_datetime(df['CloseTime'],unit ='s')
    df.set_index('CloseTime',inplace=True)
    #print(data.json())
    return df
```

```python
last_week = (pd.Timestamp.now() - pd.offsets.Day(7))

#time dayte now
print(pd.Timestamp.now())
# day we want to get back
print(pd.offsets.Day(7))
# 7 days before
print(last_week)
```

```python
btc = last_week_price('btc','bitstamp',after=last_week)
eth = last_week_price('eth','bitstamp',after=last_week)
```
